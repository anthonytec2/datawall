```python
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df=pd.read_csv('full_ml_dataset.csv')
```

```python
df.head(10)
```

```python
enc = LabelEncoder()
enc.fit(df.type.values)
```

```python
df.type=enc.transform(df.type)
```

```python
df.nameDest=df.nameDest.apply(lambda x: hash(x))
df.nameOrig=df.nameOrig.apply(lambda x: hash(x))
```

```python
df.head(10)
```

```python
y=df.isFraud.values
```

```python
X=df.iloc[:,0:9].values
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=3, test_size=0.5, random_state=0)
sss.get_n_splits(X, y)
X_train=[]
X_test=[]
y_train=[]
y_test=[]
for train_index, test_index in sss.split(X, y):
    X_train.append(X[train_index])
    X_test.append(X[test_index])
    y_train.append(y[train_index])
    y_test.append(y[test_index])
```

```python
res=[X_train[0], X_test[0], y_train[0], y_test[0]]
res2=[X_train[1], X_test[1], y_train[1], y_test[1]]
res3=[X_train[2], X_test[2], y_train[2], y_test[2]]
```

```python
pickle_out = open("Citi.pb","wb")
pickle.dump(res, pickle_out)
pickle_out.close()
pickle_out = open("JPM.pb","wb")
pickle.dump(res2, pickle_out)
pickle_out.close()
pickle_out = open("Boa.pb","wb")
pickle.dump(res3, pickle_out)
pickle_out.close()
```
